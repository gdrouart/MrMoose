```python
import numpy as np

def sync_law(x,param,redshift):
    norm, alpha = param
    # moving model in the observed frame
    x_z = x / (1. + redshift)
    y_z = (x_z**alpha)*(10**norm)*(1. + redshift)
    return y_z

# fake data
xdata=10**np.linspace(7,10,5)
norm=1.
alpha=-1.
ydata_tmp=sync_law(xdata,[norm,alpha],0.0)
yerr=ydata/5.
ydata=ydata_tmp+ydata_tmp*np.random.normal(loc=0.0,scale=0.2,size=5)

# plot for check
%matplotlib inline
import matplotlib.pyplot as plt
plt.errorbar(xdata,ydata_tmp,yerr,marker='x',label='original')
plt.plot(xdata,ydata,marker='D',label='noisy')
plt.xscale('log');plt.yscale('log')
plt.xlabel('Freq [Hz]');plt.ylabel('Flux [Jy]')
plt.legend()
```
